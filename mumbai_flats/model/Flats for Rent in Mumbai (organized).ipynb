### Data Cleaning

```python
# Dropping columns which have more than 20% null values

null_columns = df.isnull().sum()/len(df)*100
null_columns
```

```python
drop_null = null_columns[null_columns > 20].keys()
drop_null
```

```python
df_1 = df.copy()
df_1 = df_1.drop(columns = drop_null,axis=1)
df_1.head(2)
```

```python
# Dropping unnecessary columns
drop_less_import = [ 'city', 'desc',
         'id', 'id_string',
        'post_date', 'poster_name', 'title', 'trans',
       'type', 'url','user_type']
df_2 = df_1.drop(columns = drop_less_import )
df_2
```

```python
df_2.isnull().sum()
```

```python
df_3 = df_2.dropna()
df_3.reset_index(drop=True,inplace=True)
```

```python
df_3.floor_num.unique()
```

```python
df_3.floor_num.replace([-1,-2],[1,2],inplace=True)
```

### Feature Engineering

```python
def type_of_building (x):
    if x <= 8 :
        return('Buiding')
    elif x<= 40 :
        return('Tower')
    else:
        return('Skyscraper')
```

```python
df_4 = df_3.copy()
```

```python
df_4['Type_of_Apartment'] = pd.DataFrame(df_4['floor_count'].apply(type_of_building))
```

```python
df_4.sample(3)
```

```python
df_4 = df_4.drop('floor_count',axis=1)
```

```python
df_4.head()
```

```python
location = df_4.locality.value_counts()
location
```

```python
location_less_than_10 = location[location <= 10]
location_less_than_10
```

```python
df_4.locality = df_4.locality.apply(lambda x : 'other' if x in location_less_than_10 else x)
```

```python
df_5 = df_4.copy()
df_5.head()
```

### Removing Outlier

```python
df_5['price_per_sqft'] = df_5['area']/df_5['price']
df_5.sample(5)
```

```python
df_6 = df_5[~(df_5.area/df_5.bedroom_num < 200)]
df_6.shape
```

```python
upper_limit = df_6.price_per_sqft.mean() + 3*df_6.price_per_sqft.std()
upper_limit
```

```python
df_6[(df_6.price_per_sqft > upper_limit) | (df_6.price_per_sqft < lower_limit)]
```

```python
df_7 = df_6[(df_6.price_per_sqft < upper_limit) & (df_6.price_per_sqft > lower_limit)]
df_7.reset_index(drop=True)
```

```python
df_8 = df_7[df_7['bathroom_num']<df_7['bedroom_num']+2]
df_8.reset_index(drop=True,inplace=True)
df_8
```

```python
def furnishing_1 (x):
    if x == 'Unfurnished':
        return 0
    elif x == 'Semi-Furnished':
        return 1
    else:
        return 2
    
    
    
    
def Apartment (x):
    if x == 'Buiding':
        return 0
    elif x == 'Tower':
        return 1
    else:
        return 2
    
    
```

```python
df_8['furnishing'] = df_8['furnishing'].apply(furnishing_1)
df_8['Type_of_Apartment'] = df_8['Type_of_Apartment'].apply(Apartment)
```

```python
def predict_price(location,area,bedroom,floor_num,furnishing,apartment): 
    
    loc_index = np.where(X.columns==location)[0][0]

    x = np.zeros(len(X.columns))
    x[0] = area
    x[1] = bedroom
    x[2] = floor_num
    x[3] = furnishing
    x[4] = apartment
    if loc_index >= 0:
        x[loc_index] = 1

    return lr_1.predict([x])[0]
```
